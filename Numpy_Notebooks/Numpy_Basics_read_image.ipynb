# Images as NumPy Arrays: Structure and Properties

$$
\text{Image} \in \mathbb{R}^{H \times W \times C}
$$

where:
- $H$ = height (number of rows)
- $W$ = width (number of columns)
- $C$ = number of color channels (RGB $\rightarrow C = 3$)

```python
# ============================
# Images as NumPy Arrays
# Extended Inspection
# ============================

from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# Step 1: Image path
# ----------------------------
filename = "images/sedali_image.jpeg"

# ----------------------------
# Step 2: Load image
# ----------------------------
img = Image.open(filename)

# ----------------------------
# Step 3: Convert to NumPy array
# ----------------------------
img_array = np.array(img)

# ----------------------------
# Step 4: Basic properties
# ----------------------------
print("Shape:", img_array.shape)        # (H, W, C)
print("Number of dimensions:", img_array.ndim)
print("Total number of elements:", img_array.size)
print("Data type:", img_array.dtype)

# ----------------------------
# Step 5: Image geometry
# ----------------------------
height, width, channels = img_array.shape
print("Height:", height)
print("Width:", width)
print("Channels:", channels)

# ----------------------------
# Step 6: Memory usage
# ----------------------------
print("Memory usage (bytes):", img_array.nbytes)

# ----------------------------
# Step 7: Pixel access
# ----------------------------
# Top-left 5x5 pixels (RGB values)
print(img_array[:5, :5, :])

# RGB value of a single pixel (row 100, column 200)
print("Pixel (100, 200) RGB value:", img_array[100, 200])

# ----------------------------
# Step 8: Visualization
# ----------------------------
plt.imshow(img_array)
plt.axis("off")
```
