```python
from ultralytics import YOLO
import cv2

# 1️⃣ Load pretrained YOLOv8 model (nano version)
model = YOLO("yolov8n.pt")  # automatically downloads if not present

# 2️⃣ Give image path
image_path = r"C:\Neural-Networks-Files\Github-Projects\Images\yolo_3.jpg"   # <-- put your image here

# 3️⃣ Run inference
results = model(image_path)

# 4️⃣ Show results
results[0].show()  # display image with bounding boxes

# 5️⃣ Save results (optional)
results[0].save(filename=r"C:\Neural-Networks-Files\Github-Projects\Images\yolo_3_out.jpg")

# 6️⃣ Print detected objects
for box in results[0].boxes:
    cls_id = int(box.cls[0])
    conf = float(box.conf[0])
    print(f"Detected: {model.names[cls_id]} with confidence {conf:.2f}")
```
