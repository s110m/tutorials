```python
from flair.data import Sentence
from flair.models import SequenceTagger

tagger = SequenceTagger.load("ner")   # pretrained NER
sentence = Sentence("Barack Obama was born in Hawaii. Ali is a good man. We love him. He is from Iran. He study computer science in Malardalen University.")
tagger.predict(sentence)
print(sentence.get_spans("ner"))
```
