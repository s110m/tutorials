```python
import torch
from diffusers import StableDiffusionInstructPix2PixPipeline
from PIL import Image

# -----------------------------
# 1. Select device
# -----------------------------
device = "cuda" if torch.cuda.is_available() else "cpu"
dtype = torch.float16 if device == "cuda" else torch.float32

# -----------------------------
# 2. Load pretrained model
# -----------------------------
model_id = "timbrooks/instruct-pix2pix"

pipe = StableDiffusionInstructPix2PixPipeline.from_pretrained(
    model_id,
    torch_dtype=dtype,
    safety_checker=None  # remove this line if you want the default safety checker
)
pipe = pipe.to(device)

# Optional: memory optimization
if device == "cuda":
    pipe.enable_attention_slicing()

# -----------------------------
# 3. Load input image
# -----------------------------
input_image_path = "sedali_image.jpeg"   # put your image here
image = Image.open(input_image_path).convert("RGB")

# Resize to a size the model handles well
image = image.resize((512, 512))

# -----------------------------
# 4. Prompt for editing
# -----------------------------
# prompt = (
#     "clean-shaven man with smooth skin, natural face, preserve identity, "
#     "natural face, preserve cloths, same lighting and background"
# )
# prompt = (
#     "the same person 40 years older, elderly version of the same face, "
#     "realistic aging, gray hair, natural wrinkles, aged skin texture, "
#     "preserve identity, same pose, same lighting, same background, "
#     "photorealistic portrait"
# )

# negative_prompt = (
#     "young face, child, cartoon, distorted face, unrealistic skin, "
#     "blurry, extra face, deformed features"
# )

# prompt = (
#     "female version of the same person, transform the face into a woman, "
#     "feminine facial features, smooth skin, longer hair, soft jawline, "
#     "natural makeup, preserve identity, same pose, same lighting, "
#     "photorealistic portrait"
# )

# negative_prompt = (
#     "male face, beard, mustache, facial hair, distorted face, "
#     "blurry, unrealistic, extra face"
# )

prompt = (
    "female version of the same person, realistic woman face, "
    "long flowing hair, feminine facial features, soft jawline, "
    "smooth skin, natural makeup, photorealistic portrait, "
    "preserve identity, same pose, same lighting, same background"
)

negative_prompt = (
    "male face, beard, mustache, facial hair, short hair, "
    "distorted face, unrealistic, blurry"
)

# -----------------------------
# 5. Run image editing
# -----------------------------
result = pipe(
    prompt=prompt,
    image=image,
    num_inference_steps=30,
    image_guidance_scale=1.5,
    guidance_scale=7.5
).images[0]

# -----------------------------
# 6. Save result
# -----------------------------
result.save("output_woman_hair.png")
print("Saved: output_without_beard.png")
```
